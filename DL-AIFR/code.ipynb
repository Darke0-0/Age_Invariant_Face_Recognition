```python
import pandas as pd
import cv2
import numpy as np
import os
```

```python
df = pd.read_csv('CACD.csv')
```

```python
df.head()
```

```python
for i in range(len(df)):
    img = 'CACD2000/CACD01/' + df.filename[i]
    img = cv2.imread(img)
    idx = df.filename[i].replace('.',' ').replace('_',' ').split()[-2]
    dirname = df.fold[i] + '/' + 'im' + str(df['identity'][i])
    # print(dirname)
    if not os.path.exists(dirname):
        os.makedirs(dirname)  

    loc = str(idx) + '_' + str(df.age[i]) + '.jpg'
    try:
        cv2.imwrite(os.path.join(dirname,loc),img)
    except:
        continue
```
